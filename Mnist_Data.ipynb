```python
import tensorflow_datasets as tfds
mnist=tf.keras.datasets.mnist

# Create MNIST builder
builder = tfds.builder('mnist')

# Download the dataset
builder.download_and_prepare()

# Load MNIST dataset
train_dataset = builder.as_dataset(split='train', shuffle_files=True)
test_dataset = builder.as_dataset(split='test', shuffle_files=True)
```

```python
(training_images, training_labels), (test_images, test_labels) = mnist.load_data()
training_images=training_images.reshape(60000, 28, 28, 1)
training_images=training_images / 255.0
test_images = test_images.reshape(10000, 28, 28, 1)
test_images=test_images/255.0
```

```python
num_classes = 10
training_labels = tf.keras.utils.to_categorical(training_labels, num_classes)
test_labels = tf.keras.utils.to_categorical(test_labels,num_classes)
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(32, (3,3), activation='relu', input_shape=(28, 28, 1)),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(10, activation='softmax')
])
```

```python
# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# Train the model, early stopping if desired accuracy is reached
history = model.fit(training_images, training_labels, epochs=20, validation_split=0.2, callbacks=[tf.keras.callbacks.EarlyStopping(monitor='val_accuracy', patience=2)])
```

```python
test_loss, test_acc = model.evaluate(test_images, test_labels)
```
